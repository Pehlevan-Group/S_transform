```python
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
from matplotlib.patches import Rectangle
import os
if not os.path.exists('figures'):
    os.makedirs('figures')

# Set the font size to 14 for better visibility:
plt.rcParams.update({'font.size': 14})
```

```python
# Create a larger figure to accommodate the larger square
plt.figure(figsize=(8, 6))

# Create a Venn diagram
venn = venn2(subsets=(1, 1, 1), set_labels=('X', ''))

id_to_label = {
    '10': 'Var$_X$',
    '01': 'Var$_$',
    '11': 'Var$_{X,}$'
}

id_to_color = {
    '10': 'orange',
    '01': 'grey',
    '11': 'crimson'
}

for region in ['10', '01', '11']:
    venn.get_patch_by_id(region).set_color(id_to_color[region])
    venn.get_label_by_id(region).set_text(id_to_label[region])
    venn.get_patch_by_id(region).set_edgecolor('black')


# Adding and positioning the rectangle for Bias^2
rectangle = Rectangle((-0.65, 0.5), 0.3, 0.2, color="forestgreen", ec="black", clip_on=False)
plt.gca().add_patch(rectangle)
plt.text(-0.5, 0.6, "Bias$^2$", fontsize=12, ha='center', va='center')

plt.savefig('figures/LR_BV.pdf', bbox_inches='tight')
```

```python
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
from matplotlib_venn import venn3
from matplotlib.patches import Rectangle

# Set the font size to 14 for better visibility:
plt.rcParams.update({'font.size': 14})

# Create a larger figure to accommodate the modifications
plt.figure(figsize=(8, 6))

# Create a Venn diagram with three sets
venn = venn3(subsets=(5, 5, 5, 5, 5, 5, 5), set_labels=('X', '', 'F'))

id_to_label = {
    '100': 'Var$_X$',
    '010': 'Var$_$',
    '001': 'Var$_F$',
    '110': 'Var$_{{X,}}$',
    '101': 'Var$_{{X,F}}$',
    '011': 'Var$_{{,F}}$',
    '111': 'Var$_{{X,,F}}$'
}

id_to_color = {
    '100': 'orange',
    '010': 'grey',
    '001': 'purple',
    '110': 'crimson',
    '101': 'pink',
    '011': 'grey',
    '111': 'coral'
}

# Coloring and labeling the groups:
for region in ['100', '010', '001', '110', '101', '011', '111']:
    venn.get_patch_by_id(region).set_color(id_to_color[region])
    venn.get_label_by_id(region).set_text(id_to_label[region])
    venn.get_patch_by_id(region).set_edgecolor('black')

# Adjusting labels for better fit and visibility
for label in venn.subset_labels:
    label.set_fontsize(13)
    label.set_verticalalignment('center')
    label.set_horizontalalignment('center')

# Adding and positioning the rectangle for Bias^2
rectangle = Rectangle((-0.8, 0.5), 0.3, 0.2, color="forestgreen", ec="black", clip_on=False)
plt.gca().add_patch(rectangle)
plt.text(-0.65, 0.6, "Bias$^2$", fontsize=12, ha='center', va='center')

plt.savefig('figures/LRF_BV.pdf', bbox_inches='tight')

```
